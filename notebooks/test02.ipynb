Testing for file paths for ReadFile - Understanding the `resolve_path` function

```python
from pathlib import Path

path = Path.cwd()
print(path)
print(path.absolute())
```

```python
print(path.resolve())
```

```python
pat = "main.py"
pat = Path(pat)
if pat.is_absolute():
    print("This is an absolute path")
    print(pat)
    print(pat.resolve())
else:
    print("This is a relative path")
    print(pat)
    print(Path.cwd() / pat)
```
